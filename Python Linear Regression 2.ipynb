```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv(r"D:\Fingertips\Python\Project\project2\Python_Linear_Regres2.csv")
df
```

```python
df.columns
```

```python
df.describe()
```

```python
df = df.drop(['Address','Date','Postcode','YearBuilt','Lattitude', 'Longtitude'],axis=1)
df
```

```python
df.isnull().sum()
```

```python
df[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']] = df[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']].fillna(0)
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Method'].unique()
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
X = df.drop('Price',axis=1)
Y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=5)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
regression_model = LinearRegression()
```

```python
regression_model.fit(X_train,Y_train)
```

```python
predicted_price = regression_model.predict(X_test)
predicted_price
```

```python
regression_model.score(X_train,Y_train)
```

```python
regression_model.score(X_test,Y_test)
```

```python
mse = mean_squared_error(Y_test,predicted_price)
r2 = r2_score(Y_test,predicted_price)
```

```python
print('Slope:',regression_model.coef_)
print('Intercept:',regression_model.intercept_)
print('Root Mean Square Eroor:',mse)
print('R2 Score:',r2)
```

```python
ridge_reg = Ridge(alpha=50,max_iter=50)
ridge_reg.fit(X_train,Y_train)
```

```python
ridge_reg.score(X_train,Y_train)
```

```python
ridge_reg.score(X_test,Y_test)
```
